```python
import pandas as pd
```

```python
df = pd.read_csv('out-2023-08-24.csv')
```

```python
df.head
```

```python
decr_df = df[df['s1'] < 0]
```

```python
decr_df[:30]
```

```python
daily_agg = df.groupby('dt').agg({'s1':'sum', 's2':'sum'}).reset_index()
```

```python
daily_agg
```

```python
s1_dev = daily_agg['s1'].std()
s2_dev = daily_agg['s2'].std()
```

```python
s1_dev
```

```python
daily_agg['s1_norm'] = daily_agg['s1'].apply(lambda x: 1.0 if x > 3*s1_dev else x/(3*s1_dev))
daily_agg['s2_norm'] = daily_agg['s2'].apply(lambda x: 1.0 if x > 3*s2_dev else x/(3*s2_dev))
```

```python
daily_agg
```

```python
daily_agg['s1_ma10'] = daily_agg['s1_norm'].rolling(window=10).mean()
daily_agg['s1_ma30'] = daily_agg['s1_norm'].rolling(window=30).mean()
daily_agg['s1_ma60'] = daily_agg['s1_norm'].rolling(window=60).mean()
```

```python
daily_agg['s2_ma10'] = daily_agg['s2_norm'].rolling(window=10).mean()
daily_agg['s2_ma30'] = daily_agg['s2_norm'].rolling(window=30).mean()
daily_agg['s2_ma60'] = daily_agg['s2_norm'].rolling(window=60).mean()
```
